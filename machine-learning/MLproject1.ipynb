```python
import pandas as pd
```

```python
df= pd.read_csv("PS_20174392719_1491204439457_log.csv")
```

```python
df
```

```python
df["type"].unique()
```

```python
#one hot coding
#text feature to 5 numerical features
# Get one hot encoding of columns B
one_hot = pd.get_dummies(df['type'])
# Drop column B as it is now encoded
df = df.drop('type',axis = 1)
# Join the encoded df
df = df.join(one_hot)
```

```python
df = df.drop(columns=["nameOrig"])
```

```python
df["nameDest_type"]=df["nameDest"].str[:1]
df
```

```python
df = df.drop(columns=["nameDest"])
```

```python
#one hot coding
#text feature to 5 numerical features
# Get one hot encoding of columns B
one_hot = pd.get_dummies(df['nameDest_type'])
# Drop column B as it is now encoded
df = df.drop('nameDest_type',axis = 1)
# Join the encoded df
df = df.join(one_hot)
```

```python
df
```

```python
df = df.drop(columns=["M"])
```

```python
def norm_column(df, col):
    df_col = df[col]
    min_val = df_col.min()
    max_val = df_col.max()
    df[col] = (df_col - min_val) / (max_val - min_val)
```

```python
norm_cols=[
    "step",
    "amount",
    "oldbalanceOrg",
    "newbalanceOrig",
    "oldbalanceDest",
    "newbalanceDest"
]
for col in norm_cols:
    norm_column(df, col)
```

```python
df[norm_cols] = df[norm_cols].astype(float)
one_hot_cols = ['isFlaggedFraud', 'CASH_IN', 'CASH_OUT', 'DEBIT', 'PAYMENT', 'TRANSFER', 'C']
df[one_hot_cols] = df[one_hot_cols].astype(int)
```

```python
from imblearn.under_sampling import RandomUnderSampler
x = df.drop(columns=["isFraud"])
X = df[x.columns].values
print(X)
# print(X)
# x=df.columns
# x.delete(6)
# X= df[x ].values
y=df["isFraud"].values
rus = RandomUnderSampler(random_state=0)
rus.fit(X, y)
X_resampled, y_resampled = rus.fit_resample(X, y)
pass
```

```python
from sklearn.model_selection import train_test_split
X = pd.DataFrame(X_resampled, columns=x.columns)  # Ensure X_resampled is a DataFrame with the correct column names
y = pd.Series(y_resampled)  # Convert y_resampled to a Series

# Perform the train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Verify the columns of X_train
print(X_train.columns)
```

```python
import xgboost as xgb

# Create regression matrices
dtrain_reg = xgb.DMatrix(X_train, y_train, enable_categorical=True)
dtest_reg = xgb.DMatrix(X_test, y_test, enable_categorical=True)
```

```python
# Define hyperparameters

params = {"objective": "reg:squarederror", "tree_method": "hist"}
n = 100
model = xgb.train(params=params, dtrain=dtrain_reg, num_boost_round=n)
```

```python
from sklearn.metrics import mean_squared_error

preds = model.predict(dtest_reg)
```

```python
rmse = mean_squared_error(y_test, preds, squared=False)

print(f"RMSE of the base model: {rmse:.3f}")
# RMSE of the base model: 543.203
```

```python
# Create classification matrices
dtrain_clf = xgb.DMatrix(X_train, label=y_train, enable_categorical=True)
dtest_clf = xgb.DMatrix(X_test, label=y_test, enable_categorical=True)
params = {
    "objective": "binary:logistic",   # Binary classification with logistic regression
    "tree_method": "hist",            # Using histogram-based method for training
}

# Number of boosting rounds
n_rounds = 1000

# Perform cross-validation
results = xgb.cv(
    params=params,
    dtrain=dtrain_clf,
    num_boost_round=n_rounds,
    nfold=5,                        # Number of folds for cross-validation
    metrics=["logloss", "auc"],     # Metrics to evaluate during CV
    early_stopping_rounds=20,       # Stop early if no improvement in 20 rounds
    seed=42                         # Set seed for reproducibility
)

# Print the results
print(results)
```
